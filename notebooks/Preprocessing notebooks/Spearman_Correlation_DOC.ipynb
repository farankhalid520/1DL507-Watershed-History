# Spearman Correlation for DOC

In this notebook I calculate the Spearman rank correlation for Dissolved Organic Carbon (DOC) in our Lake chemistry dataset

**Author:** Muhammad Faran Khalid

```python
#imports
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Loading dataset
df = pd.read_csv('/content/drive/MyDrive/Model Data/LakeChem 2001-2022 Surface Season cleaned.csv')
```

```python
#Calculating Spearman correlation
doc_column = 'DOC (mg/l C)'

if doc_column in df.columns:
    # Calculating Spearman correlation matrix
    spearman_correlation_matrix = df.corr(method='spearman')

    # Isolating DOC correlations
    doc_correlations = spearman_correlation_matrix[doc_column].drop(doc_column)
    doc_correlations = doc_correlations.sort_values(ascending=False)

    print(doc_correlations)

    # Plotting correlations
    plt.figure(figsize=(10, 8))
    doc_correlations.plot(kind='bar')
    plt.title('Spearman Correlation with DOC')
    plt.ylabel('Spearman Correlation Coefficient')
    plt.show()
```
