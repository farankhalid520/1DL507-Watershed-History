```python
#Script by: Muhammad Faran Khalid
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Reading the CSV file into a DataFrame
file_path = '/content/drive/MyDrive/data/Lake_Chemistry_Hydrology_Data.csv'  # Replace with your CSV file path
df = pd.read_csv(file_path)
```

```python
# Filtering out rows where any of the relevant columns are empty
filtered_df = df.dropna(subset=['TOC (mg/l C)',
                                'Total water flow [m³/s]',
                                'Total station-corrected water flow [m³/s]',
                                'Total natural water flow [m³/s]',
                                'Local water flow [m³/s]'])
```

```python
# Calculating Spearman rank correlation
correlation_matrix = filtered_df[['TOC (mg/l C)',
                                  'Total water flow [m³/s]',
                                  'Total station-corrected water flow [m³/s]',
                                  'Total natural water flow [m³/s]',
                                  'Local water flow [m³/s]']].corr(method='spearman')
```

```python
# Plotting the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Spearman Rank Correlation between TOC and Various Water Flows')
plt.show()
```

```python
# Extracting Spearman correlations specifically between TOC and the water flow columns
toc_correlations = correlation_matrix['TOC (mg/l C)'].drop('TOC (mg/l C)')

# Plotting a bar graph of these correlations
plt.figure(figsize=(10, 6))
toc_correlations.plot(kind='bar', color='skyblue')
plt.title('Spearman Rank Correlation between TOC and Water Flow Parameters')
plt.xlabel('Water Flow Parameters')
plt.ylabel('Spearman Correlation Coefficient')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
